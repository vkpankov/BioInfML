```python
from Bio import Phylo
```

```python
tree = Phylo.parse('tree.txt', 'newick')
tree = list(tree)[0]
```

```python
Phylo.draw_ascii(tree)
```

```python
Phylo.draw(tree)
```

```python
tree.format = "phyloxml"

Phylo.write(tree, "converted_tree.xml", format="phyloxml")
```

```python
import ete3
from ete3 import Tree

eteTree = Tree("tree.txt", format=1)
```

```python
import random
eteTree.prune(random.sample(eteTree.get_leaves(), k =10))
```

```python
rendered = eteTree.render("ete_pruned.png", w=183, units="mm")
```

```python
from ete3 import TreeStyle
circular_style = TreeStyle()
circular_style.mode = "c"
circular_style.scale = 20
rendered = eteTree.render("ete_pruned_circular.png", w=183, units="mm", tree_style = circular_style)
```
